```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
train=pd.read_csv("dataset.csv")
```

```python
x_features=list(train.columns)
```

```python
x_features.remove('LAWYERS')
```

```python
X=train[x_features]
```

```python
cluster_new=KMeans(4)
```

```python
cluster_new.fit_predict(X)
```

```python
train['Clusters']=cluster_new.labels_
```

```python
output=train[['LAWYERS','Clusters']]
```

```python
output
```

```python
ls2=[]
for i in output['Clusters']:
    if i==0:
        ls2.append(1)
    elif i==1:
        ls2.append(2)
    elif i==2:
        ls2.append(3)
    else:
        ls2.append(4)
    
    
    
```

```python
def clu(b):
    for i in b:
        if i==0:
            print(1)
        elif i==1:
            print(2)
        elif i==2:
            print(3)
        else:
            print(4)
    
    
```
