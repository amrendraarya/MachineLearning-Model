```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
train=pd.read_csv("credit_train.csv")
```

```python
test=pd.read_csv("credit_test.csv")
```

```python
x_features=list(train.columns)
```

```python
test
```

```python
x_features.remove('Class')
```

```python
X=train[x_features]
```

```python
y=train['Class']
ls=[]
for i in y:
    if i == "Good":
        ls.append(1)
    else:
        ls.append(0)
```

```python
Y=pd.Series(ls)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
tree=LogisticRegression(penalty='l2')
```

```python
tree.fit(X_train,Y_train)
```

```python
y_predict=tree.predict(X_test)
```

```python
metrics.roc_auc_score(Y_test,y_predict)
```

```python
print(metrics.classification_report(Y_test,y_predict))
```

```python
y_features=list(test.columns)
y_features.remove("id")
```

```python
testX=test[y_features]
```

```python
testy=test['id']
testy
```

```python
prediction=tree.predict(testX)
```

```python

ls1=[]
for i in prediction:
    if i == 1:
        ls1.append("Good")
    else:
        ls1.append("Bad")
```
